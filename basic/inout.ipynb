```python
import random

hanguls = list("가나다라마바사아자차카타파하")
f = open("data/info.txt", "w", encoding="utf-8")
for i in range(500):
    name = random.choice(hanguls) + random.choice(hanguls)
    weight = random.randrange(40, 100)
    height = random.randrange(140, 200)
    f.write("{}, {}, {}\n".format(name,weight,height))

f.close()
```

```python
f = open("data/score.txt", "r", encoding="utf-8")

num = f.readline()

scores = []
while num:
    scores.append(int(num))
    num = f.readline()

f.close()

```

```python

print(scores)
print("합계 :" ,sum(scores))
print("평균 : %.2f" % (sum(scores)/len(scores)))
print("평균 : {0}".format(sum(scores)/len(scores)))
```

```python
# score 읽어서 sample.txt 쓰기

listscore = []
f = open("data/score.txt", "r", encoding="utf-8")
for num in f:
    print(num.strip())
    listscore.append(num.strip())
f.close()

f = open("data/sample.txt", "w", encoding="utf-8")
for i in listscore:
    f.write(i + "\n")
f.close()
```
