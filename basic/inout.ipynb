```python
f = open("data/score.txt", "r", encoding="utf-8")

num = f.readline()

scores = []
while num:
    scores.append(int(num))
    num = f.readline()

f.close()

```

```python

print(scores)
print("합계 :" ,sum(scores))
print("평균 : %.2f" % (sum(scores)/len(scores)))
print("평균 : {0}".format(sum(scores)/len(scores)))
```

```python
# score 읽어서 sample.txt 쓰기

listscore = []
f = open("data/score.txt", "r", encoding="utf-8")
for num in f:
    print(num.strip())
    listscore.append(num.strip())
f.close()

f = open("data/sample.txt", "w", encoding="utf-8")
for i in listscore:
    f.write(i + "\n")
f.close()
```

```python
# info.txt 읽어와서 BMI 지수 계산 후 화면 출력
# BMI = weight / ((height / 100) ** 2)
# bmi >= 25 : 과체중
# bmi >= 18.5 : 정상체중
#   : 저체중

# 나다 82 180
# 이름 : 나다
# print(f"이름 : {name}")
# 몸무게 : 82
# 키 : 180
# bmi : 33.78
# 결과 : 과체중

bmi = []

f = open("data/info.txt", "r", encoding="utf-8")
for i in f:
    print(i.strip().split(", "))
    name = i.split(", ")[0]
    weight = i.split(", ")[1]
    height = i.split(", ")[2]

    print(name , weight , height)



f.close
```

```python
f = open("data/info.txt", "r", encoding="utf-8")
for i in f:
    name = i.strip().split(", ")[0]
    weight = i.strip().split(", ")[1]
    height = i.strip().split(", ")[2]
    print(name , weight , height)
    bmi = int(weight) / ((int(height) / 100) ** 2)

    result = "";

    if bmi >= 25:
        result = "과체중"
    elif bmi >= 18.5:
        result = "정상체중"
    else:
        result = "저체중"

    print("\n".join(["이름 : {}", "몸무게 : {}", "키 : {}", "BMI : {}", "결과 : {}"]).format(name, weight, height, bmi, result))
    print()
```
