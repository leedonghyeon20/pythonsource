```python
# 엑셀 저장
import requests
import os
from openpyxl import Workbook
import re

wb = Workbook()
# 첫번째 sheet 가져오기
ws = wb.active
# 시트명 변경
ws.title = "네이버쇼핑"
# 제목행 추가
ws.append(["순위", "상품명", "가격", "주소"])
# 셀 너비 조절
ws.column_dimensions["B"].width = 100


url = "https://openapi.naver.com/v1/search/shop.json"

start=1
headers = {
    "X-Naver-Client-Id" : 'Ys3w15bcN2xSNC13UEvg',
    "X-Naver-Client-Secret" : 'aCgPAsVse3'
}


no = 1

for idx in range(10):
    # idx : 0
    start_num = start + (idx * 100)

    params = {
        "query":"마우스",
        "display":"100",
        "start":str(start_num)
    }
    res = requests.get(url, params=params, headers=headers)
    contents = res.json()
    # print(res.url)

    for item in contents['items']:
        # title b 태그 제거
        title = re.sub("<.*?>","",item['title'])
        # print(no, title, item['lprice'], item['link'])
        ws.append([no, title, item['lprice'], item['link']])
        no += 1

if not os.path.exists('./data'):
    os.makedirs('./data')

# 엑셀파일 저장
wb.save("./data/naver.xlsx")
wb.close()
```
