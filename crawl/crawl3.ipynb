```python
# 새창 띄우기
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.naver.com")
time.sleep(1)

# window.open('주소')
driver.execute_script("window.open('https://www.daum.net')")

# 브라우저에 열린 모든 창 정보 가져오기
tabs = driver.window_handles
print(tabs)

# 다음 창으로 이동
driver.switch_to.window(tabs[1])
# 다음에서 검색어 입력
element = driver.find_element(By.ID,"q")
element.send_keys("아이폰")
element.send_keys(Keys.ENTER)
time.sleep(2)

# 다음 창으로 이동
driver.switch_to.window(tabs[0])
# 네이버에서 검색어 입력
element = driver.find_element(By.ID,"query")
element.send_keys("축신두")
element.send_keys(Keys.ENTER)


time.sleep(2)
# driver.close()
driver.quit() # 창 전체 닫기
```

```python
# 스크롤
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome()
driver.maximize_window()
driver.get("https://shopping.naver.com/ns/home")
time.sleep(1)

# 검색창 찾기
element = driver.find_element(By.CLASS_NAME, "_searchInput_search_text_83jy9")
# 검색어 입력
element.send_keys("sm7b")
element.send_keys(Keys.ENTER)
time.sleep(2)

# 스크롤 이동
# 현재 문서 높이
prev_height = driver.execute_script("return document.body.scrollHeight")
while True:
    # 스크롤 이동
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    time.sleep(2)

    # 스크롤이 진행된 후 문서 높이
    curr_height = driver.execute_script("return document.body.scrollHeight")
     
    if curr_height == prev_height:
      break

    prev_height = curr_height   
    
# 스크롤 이동이 완료된 후 맨 처음 스크롤 이동
driver.execute_script("window.scrollTo(0,0)")



driver.close()
```

```python
# 검색어 넣은 후 이미지 결과 이동
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

from urllib.request import urlretrieve

driver = webdriver.Chrome()
driver.maximize_window()

driver.get("http://www.naver.com/")

# 검색창에 검색어 입력 후 엔터 
# 검색창 찾기
element = driver.find_element(By.ID, "query")
# 검색어 입력
element.send_keys("earl sweatshirt")
element.send_keys(Keys.ENTER)
time.sleep(3)

element = driver.find_element(By.XPATH, '//*[@id="lnb"]/div[1]/div/div[1]/div/div[1]/div[3]/a')
element.click()
time.sleep(1)

# 스크롤 이동
# 현재 문서 높이
prev_height = driver.execute_script("return document.body.scrollHeight")
while True:
    # 스크롤 이동
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    time.sleep(2)

    # 스크롤이 진행된 후 문서 높이
    curr_height = driver.execute_script("return document.body.scrollHeight")
     
    if curr_height == prev_height:
      break

    prev_height = curr_height   
# 스크롤 이동이 완료된 후 맨 처음 스크롤 이동
driver.execute_script("window.scrollTo(0,0)")
time.sleep(1)


# 파싱

# 작은 이미지 요소 가져오기
images = driver.find_elements(By.CSS_SELECTOR, ".image_group .thumb .link_thumb")

cnt = 1
for img in images: 
  # 작은 이미지 클릭
  img.click()
  # 오른쪽 큰 이미지 뜨도록 기다리기
  time.sleep(2)

  # 오른쪽 큰 이미지 저장
  big_img = driver.find_element(By.CSS_SELECTOR, ".viewer_image img")
  src = big_img.get_attribute("src")
  # https://search.pstatic.net/sunny/?src=https%3A%2F%2Fi.scdn.co%2Fimage%2Fab6761610000e5ebd5183ad16c1eae15a1c378c7&type=sc960_832

  # 이미지 다운로드
  urlretrieve(src, "./download/"+str(cnt)+".png")
  cnt += 1


time.sleep(3)
driver.close()

```
