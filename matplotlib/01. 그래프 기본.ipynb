### 기본 차트 생성

><b>리스트를 그래프로 나타내기</b>

```python
x = [1,2,3,4,5]
y = [12,43,25,15,550]

g = plt.plot(x, y)
# plt.show()
```

><b>파이썬 딕셔너리를 그래프로 나타내기</b>

```python
dict1 = {
    "data_x" : [1,2,3,4,5],
    "data_y" : [2,3,95,620,80]
}

plt.plot("data_x","data_y",data=dict1)
```

### 타이틀 설정

```python
plt.title("꺾은 선 차트")
g = plt.plot("data_x","data_y",data=dict1)
```

```python
x = [1,2,3,4,5]
y = [12,-43,-25,15,550]

g = plt.plot(x, y)
```

### 그리드

**전체설정**

```python
x = [1,2,3,4,5]
y = [12,-43,-25,15,550]

plt.grid(True)
g = plt.plot(x, y)
```

**x축설정**

```python

plt.grid(True, axis='x')
g = plt.plot(x, y)
```

**y축설정**

```python

plt.grid(True, axis='y')
g = plt.plot(x, y)
```
