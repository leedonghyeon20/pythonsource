### 1) 막대차트(기본)

```python
# plt.plot() : 선차트 

labels = ['강백호','서태웅','정대만']
values = [175,187,184]

g = plt.bar(labels, values)
```

><b>색상</b>

```python
g = plt.bar(labels, values, color='r')
```

```python
g = plt.bar(labels, values, color=['r','g','b'])
```

```python
g = plt.bar(labels, values, color=['r','g','b'], alpha=0.4)
```

><b>축 레이블 조정</b>

```python
plt.ylim(175,190)
g = plt.bar(labels, values)
```

><b>막대 너비 조정</b>

```python
plt.figure(figsize=(5,3))
g = plt.bar(labels, values, width=0.3)
```

><b>축 레이블 rotation 설정</b>

```python
plt.figure(figsize=(5,3))
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45)
g = plt.yticks(rotation=45)
```

><b>막대 </b>

```python
plt.bar(labels, values)
ㅎ = plt.xticks(labels, ['1번학생','2번학생','3번학생'])
```

### 2) 막대차트(심화)

**수평막대**

```python
a = plt.barh(labels, values)
```

```python
plt.barh(labels, values)
g = plt.xlim(170,195)
```

**막대 패턴**(https://matplotlib.org/stable/gallery/shapes_and_collections/hatch_style_reference.html)

```python
g = plt.bar(labels, values)

# print(g[0])
g[0].set_hatch('/')
g[1].set_hatch('x')
g[2].set_hatch('*')
```

```python

g = plt.bar(labels, values)
plt.ylim(170,195)

for idx, rec in enumerate(g):
    plt.text(idx, rec.get_height()+0.2, values[idx], ha='center', color='black')
```
