```python

import matplotlib.pyplot as plt
import koreanize_matplotlib
import pandas as pd
import numpy as np
```

### 산점도 그래프(scatter)

- x, y 상관관계를 나타낼 때 사용

```python
df = pd.read_csv("./data/student.csv")
df
```

#### 학년 컬럼 추가

```python
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

#### 기본 scatter

```python
plt.scatter(df['영어'],df['수학'])
plt.xlabel('영어')
plt.ylabel('수학')
```

- 이 그래프의 목적은 영어 점수와 수학 점수가 관계가 있느냐 이런 걸 보고 싶은 것

#### 원 크기 변경 - 랜덤 값 이용

```python
np.random.seed(42)
size = np.random.rand(8) * 1000
size
```

```python

plt.scatter(df['영어'],df['수학'],s=[size])
plt.xlabel('영어')
plt.ylabel('수학')
```

#### 원 크기 변경 - 학년 값 이용

```python
size = df['학년'] * 500
# size
plt.scatter(df['영어'],df['수학'],s=size)
plt.xlabel('영어')
g = plt.ylabel('수학')
```

#### 색상 변경

#### cmap (https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
size = df['학년'] * 500
# size
plt.scatter(df['영어'],df['수학'],s=size, c=df['학년'], cmap='viridis')
plt.xlabel('영어')
g = plt.ylabel('수학')
```

#### alpha : 투명

```python
size = df['학년'] * 500
# size
plt.scatter(df['영어'],df['수학'],s=size, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어')
g = plt.ylabel('수학')
```

#### colorbar()

```python
size = df['학년'] * 500
# size
plt.scatter(df['영어'],df['수학'],s=size, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어')
plt.colorbar()
g = plt.ylabel('수학')
```

```python
size = df['학년'] * 500
# size
plt.scatter(df['영어'],df['수학'],s=size, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어')
plt.colorbar(orientation='horizontal', ticks=[1,2,3], label='학년', shrink=0.8)
g = plt.ylabel('수학')
```

- ticks : 기본으로 나오는 눈금 대신 다른 것 지정
- label : 컬러바 라벨 지정
- shrink : 컬러바 길이 지정
- horizontal : 컬러바 위치(가로)
