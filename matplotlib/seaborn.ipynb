```python
import seaborn as sns
sns.set_theme()
```

```python

import matplotlib.pyplot as plt
import koreanize_matplotlib
import pandas as pd
import numpy as np
# from wordcloud import WordCloud
```

```python
data = {"영화제목":['크루엘라','극장판 : 귀멸의 칼날', '학교가는 길'], 
        '누적관객':[634202,1241003,149022]}

df = pd.DataFrame(data)
df
```

```python
plt.bar(x='영화제목', height='누적관객', data=df)
```

```python
sns.barplot(x='영화제목', y='누적관객', data=df)
```

```python
plt.bar(x='영화제목', height='누적관객', data=df, color=['r','g','b'])
```

```python
sns.barplot(x='영화제목', y='누적관객', data=df, palette='Set2', hue='영화제목')
```

```python
# 누적관객의 오름차순으로 바 차트 작성

sorted = df.sort_values('누적관객')
sorted
```

```python
plt.bar(x='영화제목', height='누적관객', data=sorted, color=['r','g','b'])
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='영화제목', y='누적관객', data=df, palette='Set2', hue='영화제목', order=df.sort_values('누적관객').영화제목)
```
