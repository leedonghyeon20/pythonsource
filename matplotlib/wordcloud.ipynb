```python

import matplotlib.pyplot as plt
import koreanize_matplotlib
import pandas as pd
from wordcloud import WordCloud
```

```python
text = 'square'

wc = WordCloud(background_color='white', repeat=True)
wc.generate(text)

plt.figure(figsize=(10,6))
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

```python
with open("./data/alice.txt") as f:
    text = f.read()

wc = WordCloud(background_color='white', repeat=True)
wc.generate(text)

plt.figure(figsize=(10,6))
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

```python
from PIL import Image
import numpy as np

with open("./data/alice.txt") as f:
    text = f.read()

#   
alice_mask = np.array(Image.open("./data/alice_mask.png"))



wc = WordCloud(background_color='white', repeat=True, mask=alice_mask)
wc.generate(text)

plt.figure(figsize=(10,6))
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
